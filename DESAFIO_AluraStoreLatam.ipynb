### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#total de la columna precio
ingreso_total1 = tienda["Precio"].sum()
ingreso_total2 = tienda2["Precio"].sum()
ingreso_total3 = tienda3["Precio"].sum()
ingreso_total4 = tienda4["Precio"].sum()
```

```python
print(f' el ingreso_total1 es de {ingreso_total1}')
print(f' el ingreso_total2 es de {ingreso_total2}')
print(f' el ingreso_total3 es de {ingreso_total3}')
print(f' el ingreso_total4 es de {ingreso_total4}')
```

# 2. Ventas por categoría

```python
ventas_por_cat1 = tienda.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
ventas_por_cat2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
ventas_por_cat3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
ventas_por_cat4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
```

# 3. Calificación promedio de la tienda


calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
ca, ca2, ca3, ca4 = tienda["Calificación"].mean(), tienda2["Calificación"].mean(), tienda3["Calificación"].mean(), tienda4["Calificación"].mean()
```

```python
print(f' la calificacion promedio de la tienda 1 es de {ca:.2f}')
print(f' la calificacion promedio de la tienda 2 es de {ca2:.2f}')
print(f' la calificacion promedio de la tienda 3 es de {ca3:.2f}')
print(f' la calificacion promedio de la tienda 4 es de {ca4:.2f}')
```

```python
calificacion_general = [ca, ca2, ca3, ca4]
calificacion_general
tiendas = ["tienda 1", "tienda 2", "tienda 3", "tienda 4"]
tiendas
```

```python
import matplotlib.pyplot as plt

# eje x e y
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
notas = [ca, ca2, ca3, ca4]

# line chart
plt.figure(figsize=(10, 5))

plt.plot(tiendas, notas, marker='o', linestyle='-', color='purple', linewidth=2)

plt.title('Evolución de la Calificación Promedio por Tienda')
plt.ylabel('Calificación (1 a 5)')
plt.grid(True, linestyle='--', alpha=0.5) # Agrega una cuadrícula suave de fondo

plt.ylim(0, 5)

plt.show()

```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
mas_vendidos1 = tienda.groupby("Producto")["Producto"].count().sort_values(ascending=False).head(5)
menos_vendidos1 = tienda.groupby("Producto")["Producto"].count().sort_values(ascending=True).head(5)
print(f' los productos mas vendidos en la tienda 1 son {mas_vendidos1.index.to_list()}')
print(f' los productos menos vendidos en la tienda 1 son {menos_vendidos1.index.to_list()}')
```

```python
mas_vendidos2 = tienda2.groupby("Producto")["Producto"].count().sort_values(ascending=False).head(5)
menos_vendidos2 = tienda2.groupby("Producto")["Producto"].count().sort_values(ascending=True).head(5)
print(f' los productos mas vendidos en la tienda 2 son {mas_vendidos2.index.to_list()}')
print(f' los productos menos vendidos en la tienda 2 son {menos_vendidos2.index.to_list()}')
```

```python
mas_vendidos3 = tienda3.groupby("Producto")["Producto"].count().sort_values(ascending=False).head(5)
menos_vendidos3 = tienda3.groupby("Producto")["Producto"].count().sort_values(ascending=True).head(5)
print(f' los productos mas vendidos en la tienda 3 son {mas_vendidos3.index.to_list()}')
print(f' los productos menos vendidos en la tienda 3 son {menos_vendidos3.index.to_list()}')
```

```python
mas_vendidos4 = tienda4.groupby("Producto")["Producto"].count().sort_values(ascending=False).head(5)
menos_vendidos4 = tienda4.groupby("Producto")["Producto"].count().sort_values(ascending=True).head(5)
print(f' los productos mas vendidos en la tienda 4 son {mas_vendidos4.index.to_list()}')
print(f' los productos menos vendidos en la tienda 4 son {menos_vendidos4.index.to_list()}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 20))

#tienda
plt.subplot(4, 2, 1)
plt.bar(mas_vendidos1.index, mas_vendidos1.values, color='#5D8736')
plt.title('Tienda 1: Más Vendidos')

plt.subplot(4, 2, 2)
plt.bar(menos_vendidos1.index, menos_vendidos1.values, color='red')
plt.title('Tienda 1: Menos Vendidos')

# tienda2
plt.subplot(4, 2, 3)
plt.bar(mas_vendidos2.index, mas_vendidos2.values, color='#5D8736')
plt.title('Tienda 2: Más Vendidos')

plt.subplot(4, 2, 4)
plt.bar(menos_vendidos2.index, menos_vendidos2.values, color='red')
plt.title('Tienda 2: Menos Vendidos')

# tienda3
plt.subplot(4, 2, 5)
plt.bar(mas_vendidos3.index, mas_vendidos3.values, color='#5D8736')
plt.title('Tienda 3: Más Vendidos')

plt.subplot(4, 2, 6)
plt.bar(menos_vendidos3.index, menos_vendidos3.values, color='red')
plt.title('Tienda 3: Menos Vendidos')

#tienda4
plt.subplot(4, 2, 7)
plt.bar(mas_vendidos4.index, mas_vendidos4.values, color='#5D8736')
plt.title('Tienda 4: Más Vendidos')

plt.subplot(4, 2, 8)
plt.bar(menos_vendidos4.index, menos_vendidos4.values, color='red')
plt.title('Tienda 4: Menos Vendidos')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_de_envio1 = tienda["Costo de envío"].mean()
costo_de_envio2 = tienda2["Costo de envío"].mean()
costo_de_envio3 = tienda3["Costo de envío"].mean()
costo_de_envio4 = tienda4["Costo de envío"].mean()

print(f' el costo de envio promedio de la tienda 1 es de {costo_de_envio1:.2f}')
print(f' el costo de envio promedio de la tienda 2 es de {costo_de_envio2:.2f}')
print(f' el costo de envio promedio de la tienda 3 es de {costo_de_envio3:.2f}')
print(f' el costo de envio promedio de la tienda 4 es de {costo_de_envio4:.2f}')

```
